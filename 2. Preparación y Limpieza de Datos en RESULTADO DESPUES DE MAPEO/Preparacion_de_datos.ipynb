# CAMBIAMOS LOS MESES A NUMEROS

```python
# Diccionario de meses en español a su número
meses = {
    "Enero": "01",
    "Febrero": "02",
    "Marzo": "03",
    "Abril": "04",
    "Mayo": "05",
    "Junio": "06",
    "Julio": "07",
    "Agosto": "08",
    "Septiembre": "09",
    "Octubre": "10",
    "Noviembre": "11",
    "Diciembre": "12"
}

# Convertir la columna a string para manejar tipos mixtos y evitar NaN inesperados
df["PERIODO"] = df["PERIODO"].astype(str)

# Crear una máscara para identificar valores que son años entre 2001 y 2025 (asumiendo que son strings de exactamente 4 dígitos)
mask_years = df["PERIODO"].str.match(r'^\d{4}$') & df["PERIODO"].between('2001', '2025')

# Reemplazar solo en las filas que NO son años (para evitar cambios o NaN en años)
for mes, num in meses.items():
    # Aplicar el reemplazo solo donde NO es un año
    df.loc[~mask_years, "PERIODO"] = df.loc[~mask_years, "PERIODO"].str.replace(mes, num, regex=False)

# Verificar los primeros valores para confirmar el resultado
print(df["PERIODO"].head())
```

# CREAMOS UNA  NUEVA COLUMNA QUE AGARRE LOS AÑOS EN OTRA COLUMNA

```python
anhos_validos = [str(anho) for anho in range(2001, 2026)]

# 👉 Paso 3: Definir la función que valida si un valor es un año
def es_anho(valor):
    valor_str = str(valor).strip()
    if valor_str in anhos_validos:
        return valor_str  # Devuelve el año como string
    else:
        return np.nan

# 👉 Paso 4: Aplicar la función a la columna que contiene años/meses
# Reemplaza 'NombreDeColumna' con el nombre real de la columna en tu Excel
df['anhos'] = df['PERIODO'].apply(es_anho)

# 👉 Paso 5 (opcional): Ver el resultado
print(df.head())
```

# LLENAMOS LOS NULOS CON EL AÑO QUE LES CORRESPONDE EN CADA FILA

```python
df['anhos'] = df['anhos'].ffill()
```

```python
df.head(25)
```

# UNIMOS LA  NUEVA COLUMNA PARA QUE SEA GESTION AÑO Y MES

```python
# Convertir las columnas a numéricas (forzando los errores a NaN si hay texto)
df["PERIODO"] = pd.to_numeric(df["PERIODO"], errors="coerce")
df["anhos"] = pd.to_numeric(df["anhos"], errors="coerce")

# Eliminar filas donde 'PERIODO' sea mayor o igual a 13 o NaN
df = df[df["PERIODO"].notna()]           # elimina filas con NaN en PERIODO
df = df[df["PERIODO"] < 13]              # conserva solo los valores 1–12

# Convertir finalmente a enteros
df["PERIODO"] = df["PERIODO"].astype(int)
df["anhos"] = df["anhos"].astype(int)

# Mostrar las primeras filas para verificar
df.head()
```

```python
# Renombrar columnas
df = df.rename(columns={
    "PERIODO": "Mes",
    "anhos": "Anhos"
})

# Verificar el cambio
df.head()
```
